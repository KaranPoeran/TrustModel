```python
from ema_workbench import load_results
results = load_results(r'C:/Users/karan/OneDrive/Desktop/EMA_runs/1000runs.tar.gz')
```

```python
def classify(data):
    result = data["Trust Distributor in Client"]
    classes = result[:, -1] < 0.001
    return classes
```

```python
from ema_workbench.analysis import prim
prim_obj = prim.setup_prim(results, classify, threshold=0.95, threshold_type=1)
```

```python
box_1 = prim_obj.find_box()
box_1.show_ppt()
box_1.show_tradeoff()
box_1.inspect(5, style="graph", boxlim_formatter="{: .2f}")
box_1.inspect(5)
box_1.select(5)
box_1.write_ppt_to_stdout()
box_1.show_pairs_scatter(5)
```

```python
#print prim to std_out
print(prim_obj.stats_to_dataframe())
print(prim_obj.boxes_to_dataframe())
```
